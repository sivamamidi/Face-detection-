```python
def generate_labels(xml_location):
  with open(xml_location) as xml_file:
    soup = BeautifulSoup(xml_file.read(),'xml')
    objects = soup.find_all('object')
    no_person = len(objects)

    boxes = []
    labels = []

    for object in objects:
      if object.find("name").text == "without_mask":
        labels.append(0)
      elif object.find("name").text == "mask_weared_incorrect":
        labels.append(1)
      elif object.find("name").text == "with_mask":
        labels.append(2)
      else:
        break
        
      xmin = int(object.find("xmin").text)
      xmax = int(object.find("xmax").text)
      ymin = int(object.find("ymin").text)
      ymax = int(object.find("ymax").text)

      boxes.append([xmin,ymin,xmax,ymax])

    boxes = np.array(boxes)
    labels = np.array(labels)

    target = {}
    target['labels']= labels
    target['boxes']= boxes

    return target,no_person
      






      
```

```python
import cv2
from tensorflow.keras.applications.vgg19 import preprocess_input
```

```python
image_size = 224
face_images = []
face_labels = []

for i,image_path in enumerate(images):
  image_load = cv2.imread(images_path+"/"+ image_path)

  for j in range(0,no_of_persons1[i]):
    face_loc = target_1[i]['boxes'][j]
    print(i,j)
    temp_face = image_load[face_loc[1]:face_loc[3],face_loc[0]:face_loc[2]]
    temp_face = cv2.resize(temp_face,(image_size,image_size))
    temp_face = preprocess_input(temp_face)

    face_images.append(temp_face)
    face_labels.append(target_1[i]['labels'][j])


face_images = np.array(face_images,dtype=np.float32)
face_labels = np.array(face_labels,dtype=np.float32)

print(face_images.shape,face_labels.shape)






```

```python
def show_face_and_labels(ind):
  plt.imshow(face_images[ind])
  plt.show()
  face_label = face_labels[ind]
  if face_label == 0:
    face_label_text = 'person is not wearing mask'
  elif face_label == 1:
    face_label_text = 'persin is not wearing mask properly'

  elif face_label == 2:
    face_label_text = 'person is wearing mask properly'

  return face_label_text
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
encoded_face_label = to_categorical(face_labels)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_images,test_images,train_labels,test_labels = train_test_split(face_images,encoded_face_label,test_size=0.2,stratify=encoded_face_label,random_state=20,shuffle=True)
```

```python
from tensorflow.keras import models,layers
```

```python
model = models.Sequential() # building the vgg16 architecture 
# 32 builders  
model.add(layers.Conv2D(32,(3,3),activation='relu',padding='same',input_shape=(224,224,3)))
model.add(layers.Conv2D(32,(3,3),activation='relu',padding='same'))
model.add(layers.MaxPooling2D(2,2))



# 64
model.add(layers.Conv2D(64,(3,3),activation='relu',padding='same'))
model.add(layers.Conv2D(64,(3,3),activation='relu',padding='same'))

model.add(layers.MaxPooling2D(2,2))
# 128 filters
model.add(layers.Conv2D(128,(3,3),activation='relu',padding='same'))
model.add(layers.Conv2D(128,(3,3),activation='relu',padding='same'))

model.add(layers.MaxPooling2D(2,2))
# 256 filers
model.add(layers.Conv2D(256,(3,3),activation='relu',padding='same'))
model.add(layers.Conv2D(256,(3,3),activation='relu',padding='same'))
model.add(layers.MaxPooling2D(2,2))


model.add(layers.Conv2D(512,(3,3),activation='relu',padding='same'))
model.add(layers.Conv2D(512,(3,3),activation='relu',padding='same'))

model.add(layers.MaxPooling2D(2,2))


#flatteniing the layers
model.add(layers.Flatten())

model.add(layers.Dense(512,activation='relu'))
model.add(layers.Dense(1024,activation='relu'))
model.add(layers.Dense(3,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x=train_images,y=train_labels,batch_size=100,epochs=2,validation_data=(test_images,test_labels))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy Curve')
plt.legend(['Training','Testing'],loc='best')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend(['Training','Testing'],loc='best')
```
